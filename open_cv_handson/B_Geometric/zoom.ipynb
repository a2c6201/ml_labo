```python
# 1 読み込み 2 情報 3 サイズ変更 4 チェック
import cv2
import matplotlib.pyplot as plt #Python 用 2D 描画ライブラリ
import ipywidgets.widgets as widgets

# 元の画像を読み込み
img = cv2.imread('yahboom.jpg') 
# 画像サイズを出力 print
(img. shape)
# 画像の高さと幅をそれぞれ x と y に割り当てます
x, y = img.shape[0:2]

# 元の画像を表示します
# cv.imshow('OriginalPicture', img)

# 元の半分、つまり出力にズームしますサイズの形式は (幅, 高さ) です
img_test1 = cv2.resize(img, (int(y / 2), int(x / 2)))
# cv2.imshow('resize0', img_test1)
# cv2.waitKey()

#最近傍補間によるズーム
# 元の 4 分の 1 にズーム
img_test2 = cv2.resize(img, (0, 0), fx=0.25, fy=0.25, interpolation=cv2.INTER_NEAREST) # cv.imshow 
('resize1', img_test2)
# cv.Keywait()
# cv.destroyAllWindows( )
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
dst1 = cv2.cvtColor(img_test1, cv2.COLOR_BGR2RGB)
dst2 = cv2.cvtColor(img_test2, cv2.COLOR_BGR2RGB)

# 元画像を表示
plt.imshow(img)
plt.show ()
```
