## 3.1.3 3次関数を描く
f(x) = (x-2)x (x+2)

```python
# リスト3-2-1
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return (x - 2) * x * (x + 2)
```

```python
print(f(1))
```

```python
a = np.array([1, 2, 3])

print(f(a))
type(a)
```

### 描画する範囲を決める

```python
x = np.arange(-3, 3.5, 0.5)
print(x)

# linspace(n1, n2, n)を使うと、等間隔にn分割する
# linspaceの方がn2にyを含められて、かつグラフの細かさもnでコントロールできる
# roundで四捨五入
x = np.linspace(-3, 3, 10)
print(np.round(x, 2))
```

### グラフを描画する

```python
plt.plot(x, f(x))
plt.show()
```

## もう少し見やすくしていく

```python
# 3-2-7

def f2(x):
  return (x-1) * x * (x+2)

x = np.linspace(-3, 3, 100)

# グラフ描画
plt.plot(x, f(x), 'black', label='$f$')   # グラフを登録する
plt.plot(x, f2(x), 'cornflowerblue', label='$f_2$')
# 判例表示 どれがどの関数のグラフか
plt.legend(loc='upper left')  # 判例の位置をlocで決める
# タイトル
plt.title('comparison of $f$ and $f_2$')
# ラベル
plt.xlabel('x')
plt.ylabel('y')
# 軸の範囲
plt.xlim(-3, 3)
plt.ylim(-15, 15)
plt.grid()
plt.show() # jupyter notebookではなくても使える
```

## グラフを複数並べる場合

```python
# plt.figure(figsize=(w, h))  wで横の長さ、hで縦の長さ
plt.figure(figsize=(10, 3))  # 描写領域の広さを指定
plt.subplots_adjust(wspace=0.5, hspace=0.5)  #  グラフ同士の間隔を指定
for i in range(6):
    plt.subplot(2, 3, i + 1)
    plt.title(i + 1)
    plt.plot(x, f(x), 'black')
    plt.ylim(-10 - 2 * i, 10 + 2 * i)
    plt.grid()

plt.show
```

# 三次元のグラフを描く
## 二つの変数の関数
f(x0, x1) = (2x0**2 + x1**2)exp(-(2x0**2 + x1**2))

```python
# 3-3-1
# %matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f3(x0, x1):
    ans = (2 * x0 ** 2 + x1 ** 2) * np.exp(-2 * x0 ** 2 + x1 ** 2)
    return ans

x0_n, x1_n = 9, 9
x0 = np.linspace(-2, 2, x0_n)
x1 = np.linspace(-2, 2, x1_n)
y = np.zeros((x1_n, x0_n))
for i0 in range(x0_n):
    for i1 in range(x1_n):
        y[i1, i0] = f3(x0[i0], x1[i1])  # f3を求めyに入れる
```

```python
np.set_printoptions(suppress=True)
# print(x0)
print(np.round(y, 1))
# print('y[1, 8]:', y[1, 1])
```
