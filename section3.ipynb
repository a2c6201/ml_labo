# 三次元のグラフを描く　(リスト3-3)
## 二つの変数の関数
f(x0, x1) = (2x0**2 + x1**2)exp(-(2x0**2 + x1**2))

```python
# リスト3-3
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f3(x0, x1):
    ans = (2 * x0 ** 2 + x1 ** 2) * np.exp(-(2 * x0 ** 2 + x1 ** 2))
    return ans

x0_n, x1_n = 9, 9
x0 = np.linspace(-2, 2, x0_n)   # [-2, -1.5, -1.0, -0.5, 0, 0.5, 1.0, 1.5, 2.0]
x1 = np.linspace(-2, 2, x1_n)
y = np.zeros((x1_n, x0_n))
for i0 in range(x0_n):
    for i1 in range(x1_n):
        y[i1, i0] = f3(x0[i0], x1[i1])  # f3を求めyに入れる
# 縦(行が)x1で横（列が）x0を表している
```

```python
np.set_printoptions(suppress=True)
# print(x0)
print(np.round(y, 1))
# print('y[1, 8]:', y[1, 1])
```

## 数値を色で表現する　pcolorを使う 3-3-4

## 関数の表面を面で表す　surface 3-3-5

```python
print('x0:', x0)
print('x1:', x1)

xx0, xx1 = np.meshgrid(x0, x1)

print('xx0:', xx0)
print('xx1:', xx1)
print(xx0[0, 8])

plt.figure(figsize=(5, 3.5))
ax = plt.subplot(projection='3d')   # 3Dグラフを作成 第一引数、第二引数を設定すれば複数のグラフを描画できる
ax.plot_surface(
    xx0,    # x座標
    xx1,    # y座標
    y,      # z座標
    rstride=1, # 何行おきに線を引くか
    cstride=1, # 
    alpha=0.3,      #透明度
    color='blue',
    edgecolor='black'
    )

ax.set_zticks((0, 0.3))    # z軸の目盛り指定
ax.view_init(75, -60)   # グラフの向きの指定(第一引数は上下の回転角度、第二引数は正だと時計回り、負の数だと反時計回り)

plt.show()
```

```python

# 解像度を50✖️50、rstride, cstride=5として表示すると

x0_n, x1_n = 50, 50
x0 = np.linspace(-2, 2, x0_n)   # [-2, -1.5, -1.0, -0.5, 0, 0.5, 1.0, 1.5, 2.0]
x1 = np.linspace(-2, 2, x1_n)
y = np.zeros((x1_n, x0_n))
for i0 in range(x0_n):
    for i1 in range(x1_n):
        y[i1, i0] = f3(x0[i0], x1[i1])  # f3を求めyに入れる


xx0, xx1 = np.meshgrid(x0, x1)

plt.figure(figsize=(5, 3.5))
ax = plt.subplot(projection='3d')   # 3Dグラフを作成 第一引数、第二引数を設定すれば複数のグラフを描画できる
ax.plot_surface(
    xx0,    # x座標
    xx1,    # y座標
    y,      # z座標
    rstride=5, # 何行おきに線を引くか
    cstride=5, # 
    alpha=0.3,      #透明度
    color='blue',
    edgecolor='black'
    )

ax.set_zticks((0, 0.3))    # z軸の目盛り指定
ax.view_init(75, -60)   # グラフの向きの指定(第一引数は上下の回転角度、第二引数は正だと時計回り、負の数だと反時計回り)

plt.show()
```
